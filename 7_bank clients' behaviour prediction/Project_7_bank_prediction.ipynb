# Отток клиентов

По представленным данным о клиентах «Бета-Банка» и их поведении нужно **спрогнозировать, уйдёт клиент из банка в ближайшее время или нет** (так как в последние месяцы наблюдается небольшой отток клиентов). 

Для этого требуется решить **следующие задачи**:
* подготовить данные,
* обучить модели без учёта дисбаланса в целевом признаке и рассчитать метрики качества этих моделей,
* устранить дисбаланс в данных и обучить модели,
* выбрать наиболее эффективную модель с предельно большим значением *F1*-меры; дополнительно измерить *AUC-ROC*.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

Предоставлены следующие данные.
Признаки:
* `RowNumber` — индекс строки в данных
* `CustomerId` — уникальный идентификатор клиента
* `Surname` — фамилия
* `CreditScore` — кредитный рейтинг
* `Geography` — страна проживания
* `Gender` — пол
* `Age` — возраст
* `Tenure` — сколько лет человек является клиентом банка
* `Balance` — баланс на счёте
* `NumOfProducts` — количество продуктов банка, используемых клиентом
* `HasCrCard` — наличие кредитной карты
* `IsActiveMember` — активность клиента
* `EstimatedSalary` — предполагаемая зарплата

Целевой признак:
* `Exited` — факт ухода клиента.

## Исследование задачи

Как стало понятно из исследования данных, покинули банк 2037 клиентов, что составляет менее 20% от выборки. Таким образом, мы имеем дело с несбалансированной выборкой по целевому признаку. Тем не менее, попробуем обучить ряд моделей без учёта этого дисбаланса и оценить качество их предсказаний.

Поскольку целевой признак выражается в двух категориях, то понадобятся алгоритмы классификации (в данном случае, бинарной).

Сначала подготовим данные для работы моделей.

### Подготовка данных без учёта дисбаланса классов

```python
# разобъём данные на целевой признак и остальные признаки
target = data['Exited']
features = data.drop('Exited', axis=1)

# разделим выборку на три части: обучающую, валидационную и тестовую - в соотношении 3:1:1
# сначала выделим обучающую выборку и тестово-валидационную часть
features_train, features_tv, target_train, target_tv, = train_test_split(
    features, target, test_size=0.4, random_state=12345)

# на втором этапе разделим тестово-валидационную на две части
features_valid, features_test, target_valid, target_test = train_test_split(
    features_tv, target_tv, test_size=0.5, random_state=12345)

# check
target_test.shape
```

```python
# Стандартизируем признаки с помощью StandardScaler 
numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

scaler = StandardScaler()
scaler.fit(features_train[numeric])
pd.options.mode.chained_assignment = None

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

# check
features_train.head()
```

### Обучение моделей без учёта дисбаланса классов

Начнём с алгоритма **решающего дерева**. Подберём глубину дерева, которая даст лучшее значение метрики F1 (определяющей баланс между полнотой и точностью выявления объектов "положительного" класса).

```python
for depth in range(1, 16):
    model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model.fit(features_train, target_train)
    predicted_train = model.predict(features_train)
    predicted_valid = model.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_train, predicted_train))
    print()
```

Итак, наилучшего значения F1-меры (почти 0,57) добилось дерево с глубиной 6 (к тому же эта модель явно не переобучилась). Однако такой результат не дотягивает до нужного.

Исследуем модели **случайного леса**.

```python
# возьмём глубину 6 (хорошую для одного дерева) и исследуем модели с разным количеством деревьев
for number in range(100, 1100, 100):
    model = RandomForestClassifier(random_state=12345, n_estimators=number, max_depth=6)
    model.fit(features_train, target_train)
    predicted_train = model.predict(features_train)
    predicted_valid = model.predict(features_valid)
    print("n_estimators =", number, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_train, predicted_train))
    print()
```

Лучший результат дал лес из 900 деревьев. Посмотрим, может ли улучшить результат изменение глубины деревьев.

```python
for depth in range(1, 16):
    model = RandomForestClassifier(random_state=12345, n_estimators=900, max_depth=depth)
    model.fit(features_train, target_train)
    predicted_train = model.predict(features_train)
    predicted_valid = model.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_train, predicted_train))
    print()
```

Видим, что лес из 900 деревьев глубиной 12 почти даёт нужный показатель F1-меры (0.588), но модель переучивается.

Посмотрим, какой результат даст **логистическая регрессия**.

```python
model = LogisticRegression(random_state=12345, solver='liblinear')
model.fit(features_train, target_train)
predicted_train = model.predict(features_train)
predicted_valid = model.predict(features_valid)
print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
print('F1 на обучающей выборке:', f1_score(target_train, predicted_train))
```

### Выводы - обучение моделей на несбалансированных данных

Данные были подготовлены для классификации, но из-за дисбаланса классов достичь значения метрики F1 не удалось ни одной модели. Логистическая регрессия показала неудовлетворительные результаты, а решающее дерево с глубиной 6 и случайный лес с количеством деревьев 900 и глубиной 12 продемонстрировали хорошие способности к выявлению объектов положительного класса, но также не достигли целевого значения.

## Борьба с дисбалансом

### Указание моделям на дисбаланс классов

Перед устранением дисбаланса классов целевого признака проверим, улучшится ли качество предсказаний моделей, если внести в набор параметров `class_weight='balanced'`.

```python
# решающее дерево model_tree_b
model_tree_b = DecisionTreeClassifier(random_state=12345, max_depth=6, class_weight='balanced')
model_tree_b.fit(features_train, target_train)
predicted_valid = model_tree_b.predict(features_valid)
print('F1 для дерева с учётом баланса:', f1_score(target_valid, predicted_valid))

# случайный лес model_forest_b
model_forest_b = RandomForestClassifier(random_state=12345, n_estimators=900, max_depth=12,
                              class_weight='balanced')
model_forest_b.fit(features_train, target_train)
predicted_valid = model_forest_b.predict(features_valid)
print('F1 для леса с учётом баланса:', f1_score(target_valid, predicted_valid))

# логистическая регрессия model_lr_b
model_lr_b = LogisticRegression(random_state=12345, solver='liblinear', class_weight='balanced')
model_lr_b.fit(features_train, target_train)
predicted_valid = model_lr_b.predict(features_valid)
print('F1 для логистической регрессии с учётом баланса:', f1_score(target_valid, predicted_valid))
```

Результаты всех моделей несколько улучшились, случайный лес даже достиг нужного уровня F1, но,  вероятно, есть пространство для дальнейшего улучшения.

Внесём изменения в данные (а именно - в обучающую выборку), чтобы изучить, как устранение дисбаланса повлияет на качество предсказаний. Будем учитывать, что объектов с целевым признаком "1" (то есть тех, кто расторг договор с банком) в 4 раза меньше, чем объектов с целевым признаком "0".

### Применение техник upsampling и downsampling

Сначала увеличим количество объектов с целевым признаком "1" в обучающей выборке - применим технику upsampling. Пусть эта тренировочная выборка будет обозначаться "upsampled".

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)

#check
features_upsampled.shape
```

Теперь уменьшим количество объектов с целевым признаком "0". Пусть эта обучающая выборка обозначается "downsampled".

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] 
                                     + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] 
                                   + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.4)

# check
features_downsampled.shape
```

В последнем случае обучающая выборка оказалась относительно небольшой.

Тем не менее, сравним, как работают модели на обновлённых данных и посмотрим, какую метрику F1 они дают.

### Выбор лучшей модели

#### Решающее дерево

Изучим, дерево какой глубины будет лучше работать с валидационной выборкой, если обучить его на **"upsampled"** данных. 

```python
for depth in range(1, 16):
    model_tree_up = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree_up.fit(features_upsampled, target_upsampled)
    predicted_train = model_tree_up.predict(features_upsampled)
    predicted_valid = model_tree_up.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_upsampled, predicted_train))
    print()
```

Видно, что дерево переобучается, а качество предсказаний снизилось по сравнению с обучением того же дерева на несбалансированной выборке.

Посмотрим, каков будет эффект, если взять для обучения **"downsampled" выборку**.

```python
for depth in range(1, 16):
    model_tree_dn = DecisionTreeClassifier(random_state=12345, max_depth=depth)
    model_tree_dn.fit(features_downsampled, target_downsampled)
    predicted_train = model_tree_dn.predict(features_downsampled)
    predicted_valid = model_tree_dn.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_downsampled, predicted_train))
    print()
```

На меньшей обучающей выборке решающее дерево справилось значительно лучше: на глубине 6 показатель достиг почти 0.57 (на текущий момент - лучший результат для всех моделей с алгоритмом решающего дерева), при этом не наблюдалось такого переобучения, как на больших тренировочных данных.

Посмотрим, насколько хорошо справляется эта модель с выявлением объектов с позитивным целевым признаком.

```python
model_tree_dn = DecisionTreeClassifier(random_state=12345, max_depth=7)
model_tree_dn.fit(features_downsampled, target_downsampled)

probabilities_valid = model_tree_dn.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

print('AUC-ROC для "downsampled":', roc_auc_score(target_valid, probabilities_one_valid))
```

В целом, можно говорить о том, что модель довольно эффективно выделяет объекты с положительным целевым признаком - по сравнению со случайной моделью, кривая ROC будет находиться выше.

#### Случайный лес

Исследуем два леса с количеством деревьев 100. Так можно будет понять, на какой выборке затем искать оптимальное количество деревьев.

Сначала обучим на "upsampled" данных.

```python
for depth in range(1, 16):
    model_forest_up = RandomForestClassifier(random_state=12345, 
                                             n_estimators=100, max_depth=depth)
    model_forest_up.fit(features_upsampled, target_upsampled)
    predicted_train = model_forest_up.predict(features_upsampled)
    predicted_valid = model_forest_up.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_upsampled, predicted_train))
    print()
```

Видим, что с ростом глубины деревьев качество предсказаний улучшается, на глубине 10 - лучший показатель.

Посмотрим на модель, обученную на уменьшенном количестве данных.

```python
for depth in range(1, 16):
    model_forest_dn = RandomForestClassifier(random_state=12345, 
                                             n_estimators=100, max_depth=depth)
    model_forest_dn.fit(features_downsampled, target_downsampled)
    predicted_train = model_forest_dn.predict(features_downsampled)
    predicted_valid = model_forest_dn.predict(features_valid)
    print("max_depth =", depth, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_downsampled, predicted_train))
    print()
```

Случайный лес показывает отличные результаты, если обучается на небольшой выборке. Запомним, что хорошее значение метрики F1 даёт глубина 10. Посмотрим, каковы будут результаты, если мы оставим эту глубину и попробуем увеличить количество деревьев.

```python
for number in range(200, 1100, 100):
    model_forest_dn = RandomForestClassifier(random_state=12345, 
                                             n_estimators=number, max_depth=10)
    model_forest_dn.fit(features_downsampled, target_downsampled)
    predicted_train = model_forest_dn.predict(features_downsampled)
    predicted_valid = model_forest_dn.predict(features_valid)
    print("n_estimators =", number, ":")
    print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
    print('F1 на обучающей выборке:', f1_score(target_downsampled, predicted_train))
    print()
```

Видим, что самый хороший результат по метрике F1 даёт случайный лес из 700 деревьев. Тогда целесообразно оставить **случайный лес из 700 деревьев глубиной 10** в качестве предпочтительного варианта.

Посмотрим, какова метрика AUC-ROC для такого случайного леса на валидационной выборке.

```python
model_forest_dn = RandomForestClassifier(random_state=12345, 
                                             n_estimators=700, max_depth=10)
model_forest_dn.fit(features_downsampled, target_downsampled)

probabilities_valid = model_forest_dn.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

print('AUC-ROC для "downsampled":', roc_auc_score(target_valid, probabilities_one_valid))
```

В целом, видим, что качество этой модели заметно лучше, чем случайной.

#### Логистическая регрессия

Изучим также работу логистической регрессии на увеличнной и уменьшенной сбалансированных выборках.

```python
# на upsampled выборке
model_lr_up = LogisticRegression(random_state=12345, solver='liblinear')
model_lr_up.fit(features_upsampled, target_upsampled)
predicted_train = model_lr_up.predict(features_upsampled)
predicted_valid = model_lr_up.predict(features_valid)
print('upsampled')
print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
print('F1 на обучающей выборке:', f1_score(target_upsampled, predicted_train))
print()

# на downsampled выборке
model_lr_dn = LogisticRegression(random_state=12345, solver='liblinear')
model_lr_dn.fit(features_downsampled, target_downsampled)
predicted_train = model_lr_dn.predict(features_downsampled)
predicted_valid = model_lr_dn.predict(features_valid)
print('downsampled')
print('F1 на валидационной выборке:', f1_score(target_valid, predicted_valid))
print('F1 на обучающей выборке:', f1_score(target_downsampled, predicted_train))
```

```python
probabilities_valid = model_lr_up.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

print('AUC-ROC для "upsampled":', roc_auc_score(target_valid, probabilities_one_valid))
print()

probabilities_valid = model_lr_dn.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]

print('AUC-ROC для "downsampled":', roc_auc_score(target_valid, probabilities_one_valid))
```

Вновь видим, что увеличение обучающей выборки не помогло модели улучшить качество предсказаний (в том числе, и по сравнению с применением параметра, балансирующего классы). Уменьшение обучающей выборки, напротив, помогло: удалось добиться результата, лучшего для алгоритма логистической регрессии. Хотя этот результат значительно хуже, чем у моделей, использующих другие алгоритмы.

В ходе исследования применялись три метода борьбы с дисбалансом классов целевого признака.

Что же касается логистической регрессии, то она работала даже хуже, чем после внесения параметра, указывающего на несбалансированность классов.
