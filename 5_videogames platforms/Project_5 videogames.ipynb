# Исследование продаж компьютерных игр

Представлены данные по продажам игр в различных жанрах, для различных платформ в нескольких регионах мира. **Цель проекта** - выявить закономерности, определяющие коммерческую успешность игры.

**Задачи исследования**:
1. Обработать данные, привести их к нужным типам, подсчитать суммарные продажи игр в регионах.
2. Исследовать данные:
    * определить актуальный период для проведения анализа,
    * отследить изменение продаж во времени по отдельным платформам,
    * выявить перспективные с точки зрения прибыльности платформы,
    * выявить популярность различных платформ в разных регионах,
    * определить, есть ли связь между продажами игр и отзывами на них пользователей и критиков,
    * выявить более и менее прибыльные жанры игр.
3. Составить портрет пользователя из каждого региона:
    * выявить для каждого региона наиболее популярные платформы,
    * определить 5 наиболее популярных жанров,
    * выяснить, влияет ли рейтинг ESRB (Entertainment Software Rating Board) на продажи в каждом регионе.
4. Проверить гипотезы:
    * О равенстве средних пользовательских рейтингов платформ Xbox One и PC.
    * О различии средних пользовательских рейтингов игр в жанрах Action и Sports.

**Описание данных**:
* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB.

## Обработка данных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats as st
import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('/datasets/games.csv')

display(data.head())
data.info()
```

Всего в таблице 16715 строк, 11 столбцов. 

В первую очередь, целесообразно перевести названия столбцов в нижний регистр, чтобы избежать путаницы и ошибок, а также нужно проверить, нет ли в таблице явных дубликатов. 

Затем необходимо обработать каждый из столбцов: исследовать их содержимое, заполнить пропуски, если они есть, привести данные к нужным типам.

```python
# переименуем столбцы
data = data.rename(columns={
    'Name': 'name',
    'Platform': 'platform',
    'Year_of_Release': 'year_of_release',
    'Genre': 'genre',
    'NA_sales': 'na_sales',
    'EU_sales': 'eu_sales',
    'JP_sales': 'jp_sales',
    'Other_sales': 'other_sales',
    'Critic_Score': 'critic_score',
    'User_Score': 'user_score',
    'Rating': 'rating'}
)
# check
data.columns
```

```python
# проверим на наличие дубликатов
data.duplicated().sum()
```

В столбце, содержащем название игры, есть два пропущенных значения. Посмотрим, в каких строках.

```python
data[data['name'].isna()]
```

К сожалению, по этим двум играм нет почти никакой информации, возможно, данные о них были утеряны. Восстановить информацию не представляется возможным, поэтому целесообразно удалить эти строки.

```python
# удаляем пока только строки с пропусками в столбце name
data = data.dropna(subset=['name'])
```

После этого количество выпускаемых игр стало снижаться и в 2015-16 годах находилось примерно на уровне 500-600 штук в год.

Удалим пропущенные значения, так как данных о времени выпуска нет только о 2% игр и переведём данные в целочисленный формат.

```python
data = data.dropna(subset=['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype('int')

# check
data.dtypes
```

### Столбец critic_score

В этом столбце пропущено более половины значений. Посмотрим, как распределены оставшиеся.

```python
data['critic_score'].describe()
```

```python
data['critic_score'].hist(bins=20)
```

Критики оценивают игры по шкале от 0 до 100 баллов, при этом ни одна из представленных игр не была оценена на 100. Большинство игр получают от 60 до 85 баллов, при этом более 90 баллов, как видно, получили только около 300 игр.

Вероятно, пропуски в этом столбце связаны с тем, что не все игры в принципе оценивались критиками. Заменим все пропуски на "-1".

```python
data['critic_score'] = data['critic_score'].fillna(-1)

# check
data['critic_score'].describe()
```

### Столбец user_score

В этом столбце данные представлены в формате object, значит, возможно, встречаются тектовые значения. Посмотрим, что содержится в столбце.

```python
data['user_score'].unique()
```

Так и есть, наряду с числами, в столбце мы видим: 
* nan - либо отсутствующие, либо потерянные значения,
* tbd - то есть, специально не заполненные, to be determined. Вероятно, они относятся к новым играм, по которым пользовательский рейтинг пока не определён.

```python
print(data[data['user_score'] == 'tbd']['name'].count())
print(data[data['user_score'].isna()]['name'].count())

data[data['user_score'] == 'tbd']['year_of_release'].describe()
```

Предположение не оправдалось, однако для целей настоящего исследования важны только количественные данные в столбце `user_score`. Если привести данные к типу float - что нужно для вычислений с этим столбцом, то все пропуски будут выглядеть как nan.

Заменять пропущенные значения на 0, медиану или среднее нельзя - их слишком много, и это может повлиять на выводы.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')

# check
data.info()
```

Тип данных изменён, данные о рейтинге есть только в 7590 строках - остальные 9123 неизвестны.

Можно заменить все неизвестные значения на "-1". 

```python
data['user_score'] = data['user_score'].fillna(-1)

# check
data['user_score'].describe()
```

Теперь можно узнать, как распределяются пользовательские рейтинги.

```python
data['user_score'].describe()
```

```python
data['user_score'].hist()
```

Не принимаем во внимание искусственно введённую оценку -1. Видим, что распределение оценок пользователей напоминает распределение оценок критиков, при этом среднее (если бы мы пересчитали его для 100-балльной шкалы) даже выше. Чаще всего пользователи оценивают игры на 6-8 баллов.

### Столбец rating

```python
data['rating'] = data['rating'].fillna('unknown')

# check
data['rating'].value_counts()
```

### Новый столбец - total_sales

#### Оценки пользователей и критиков

Возьмём для анализа перспективную платформу **PS4** и попробуем выявить, влияют ли оценки, которые дают играм пользователи и критики, на продажи этих игр.

```python
# проведём расчёты коэффициентов на датасете, в котором нет искусственных отрицательных значений
ac_data_corr = ac_data.query('user_score != -1 & critic_score != -1')
ac_data_corr[ac_data_corr['platform'] == 'PS4'][['total_sales', 'critic_score', 'user_score']].corr()
```

Судя по коэффициентам корреляции, продажи игр находятся в слабой положительной зависимости от оценок критиков и практически не связаны с оценками пользователей. Примечательно, что и оценки критиков и пользователей не всегда связаны.

Построим диаграммы рассеяния, чтобы визуально оценить корреляцию.

```python
x = ac_data_corr[ac_data_corr['platform'] == 'PS4']['critic_score']
y = ac_data_corr[ac_data_corr['platform'] == 'PS4']['total_sales']

plt.scatter(x, y, c='m')
plt.suptitle('Зависимость общих продаж игр на PS4 от оценок критиков')
plt.xlabel('Оценки критиков')
plt.ylabel('Общие продажи')
plt.show()
```

Действительно, можем говорить лишь об умеренной связи между оценками критиков и общими продажами игр на платформе PS4. Однако все игры, которые продавались хорошо, получили высокие оценки. И самая продаваемая игра была оценена наиболее высоко.

```python
y = ac_data_corr[ac_data_corr['platform'] == 'PS4']['total_sales']
x = ac_data_corr[ac_data_corr['platform'] == 'PS4']['user_score']

plt.scatter(x, y, c='blue')
plt.suptitle('Зависимость общих продаж игр на PS4 от оценок пользователей')
plt.xlabel('Оценки пользователей')
plt.ylabel('Общие продажи')
plt.show()
```

А в случае оценок пользователей вряд ли можно говорить о наличии зависимости: высокие оценки получило множество игр, которые продавались не очень хорошо и наоборот.

## Проверка гипотез

В данном разделе необходимо проверить, что:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Средние пользовательские рейтинги жанров Action и Sports разные.

### Проверка гипотезы о равенстве рейтингов платформ

Посмотрим, каков средний рейтинг игр, выпускаемых для платформ XOne и PC. Будем изучать данные в актуальном периоде.

Создадим две выборки, очистив данные о пользовательских рейтингах от пропущенных значений и используя данные за актуальный период.

```python
# для рейтингов игр на XOne
xone_sample = ac_data.loc[(ac_data['platform'] == 'XOne') & (ac_data['user_score'] > 0)]


#check
xone_sample.info()
```

```python
# для рейтингов игр на PC
pc_sample = ac_data.loc[(ac_data['platform'] == 'PC') & (ac_data['user_score'] > 0)]

#check
pc_sample.info()
```

```python
print('Средний пользовательский рейтинг игр на Xbox One:', xone_sample['user_score'].mean())
print('Средний пользовательский рейтинг игр на PC:', pc_sample['user_score'].mean())

# можно также обратить внимание на основные статистики оценок пользователей
print(xone_sample['user_score'].describe())
pc_sample['user_score'].describe()
```

На первый взгляд кажется, что рейтинги похожи. Примем гипотезу об их равенстве в качестве нулевой. Альтарнативной двусторонней гипотезой будет предположение о том, что они неравны. Для проверки нулевой гипотезы целесообразно использовать t-тест, так как речь идёт о сравнении средних генеральных совокупностей по представленным выборкам.

Генеральными совокупностями в данном случае представляются рейтинги игр на этих платформах за весь период. Их дисперсии мы можем принять равными. Установим уровень критической значимости на 5%.

```python
alpha = 0.05

results = st.ttest_ind(
    xone_sample['user_score'],
    pc_sample['user_score'])

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Можно констатировать, что полученные данные не противоречат гипотезе о равенстве средних рейтингов игр для Xbox One и PC.

```python
print(act_sample['user_score'].describe())
sports_sample['user_score'].describe()
```

На первый взгляд, средний рейтинг игр в жанре экшен выше, чем рейтинг спортивных симуляторов. Нам необходимо проверить, различаются ли они. 

**Примем в качестве нулевой гипотезы предположение о равенстве этих рейтингов**. Альтернативная гипотеза, в таком случае, вновь будет двусторонней и будет формулироваться таким образом: средние пользовательские рейтинги игр в жанрах экшен и спорт различаются.

Здесь также целесообразно использовать t-тест двух выборок, чтобы проверить равенство средних двух генеральных совокупностей. Предположим, что дисперссии в них примерно равны, а критический уровень статистической значимости вновь установим на уровне 5%.

```python
alpha = 0.05

results = st.ttest_ind(
    act_sample['user_score'],
    sports_sample['user_score'])

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Вероятность случайно получить данные, которые соответствовали бы нулевой гипотезе (о равенстве прользовательских рейтингов игр в двух жанрах) крайне мала (около 0,05 * 10^-22). Таким образом, нам нужно отвергнуть нулевую гипотезу и полагать, что пользовательские рейтинги игр в жанрах экшен и спорт различаются.
