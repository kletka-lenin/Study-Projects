#### Столбец `ceiling_height`

Большое количество пропусков есть в данных о высоте потолков (**`ceiling_height`**). Посмотрим более внимательно, в каких случаях пропущены данные.

```python
data.loc[data['ceiling_height'].isna()].describe()
```

Строки с пропусками не слишком отличаются от остальных данных.

Построим гистограмму: как по выборке распределена высота потолков.

```python
data['ceiling_height'].hist()
```

Из этой гистограммы не ясно ничего, за исключением того, что значения высоты потолков более 10 метров крайне редки. Изменим гистограмму в соответствии с этим выводом.

```python
data['ceiling_height'].hist(range=(1,10))
```

Стало понятно, что, в основном, высота потолков колеблется в довольно узких пределах: от 2 до 5 метров, причём даже потолки выше 4 метров встречаются довольно редко. 

Посмотрим на экстремальные наблюдения: есть ли в них что-то необычное.

```python
# сначала изучим маленькие значения, допустим, потолки не могут быть ниже 2 метров
data.query('ceiling_height < 2')
```

В этих строках нет ничего необычного, поэтому, как представляется, можно заменить эти значения на медианную высоту потолков.

Теперь рассмотрим слишком высокие потолки.

```python
data.query('ceiling_height > 5')['ceiling_height'].hist(bins=30, range=(5,100), figsize=(12,5))
```

Видно, что множество "высоких потолков" - это варианты около 25 метров. Можно предположить, что люди, которые публиковали объявление, случайно не поставили разделитель между целой и дробной частью числа. То есть, фактически, это ошибки. Их можно исправить, разделив значения больше 20 м на 10.

Остаются данные от 5 до 20 метров. Посмотрим на них отдельно. Возможно, это особые объекты недвижимости.

```python
data.query('5 < ceiling_height < 20')
```

Заметно, что несколько объектов из этого набора - большой площади. Это могут быть частные дома или двухуровневые квартиры. В таких случаях замена нецелесообразна. Однако есть и некоторые аномальные значения более 10 метров в объектах, которые выглядят совершенно обычными. 

Исходя из этого можно предложить следующую схему замен:
* если высота потолков менее 2 метров, заменяем значение на медианное (2,65 м).
* если высота потолка от 2 до 10 метров, не изменяем значение.
* если высота более 10 метров и до 20 метров (20 не включаем), то заменяем на медианное.
* если высота от 20 метров, то делим значение на 10.
* пропущенные значения не будем заменять медианным, так как их слишком много: это может отразиться на результатах исследования.

```python
# Заменим значения с помощью цикла

for item in data['ceiling_height']:
    if item < 2:
        data['ceiling_height'] = data['ceiling_height'].replace(item, 2.65)
    elif item >= 20:
        data['ceiling_height'] = data['ceiling_height'].replace(item, item / 10)
    elif item > 10:
        data['ceiling_height'] = data['ceiling_height'].replace(item, 2.65)

# проверяем, что получилось
print(data['ceiling_height'].min())
print(data['ceiling_height'].max())
data['ceiling_height'].hist(range=(1,10))
```

Получилось похоже на распределение необработанного столбца, при этом без слишком маленьких и слишком больших значений.

## Расчёты и добавление результатов в таблицу

В таблицу необходимо добавить следующие столбцы:
* цену квадратного метра (`price`),
* день недели (`pub_day`), месяц (`pub_month`) и год (`pub_year`) публикации объявления,
* этаж квартиры (`floor_category`) - первый, последний или другой,
* соотношение жилой и общей площади (`liv_to_total_ratio`), а также отношение площади кухни к общей (`kit_to_total_ratio`).

```python
# Добавляем столбец с ценой квадратного метра
data['price'] = data['last_price'] / data['total_area']

# Столбец с днём публикации
data['pub_day'] = data['first_day_exposition'].dt.weekday

# Столбец с месяцем публикации
data['pub_month'] = data['first_day_exposition'].dt.month

# Столбец с годом публикации
data['pub_year'] = data['first_day_exposition'].dt.year

# Столбец с соотношением жилой площади к общей
data['liv_to_total_ratio'] = data['living_area'] / data['total_area']

# Столбец с соотношением площади кухни к общей
data['kit_to_total_ratio'] = data['kitchen_area'] / data['total_area']
```

```python
# Создадим функцию, которая разберёт этажи на категории
def floor_category(row):
    floor = row['floor']
    total_floors = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == total_floors and floor != 1:
        return 'последний'
    else:
        return 'другой'

data['floor_category'] = data.apply(floor_category, axis=1)

# посмотрим, сработала ли функция
data['floor_category'].value_counts()
```

### Недвижимость в Санкт-Петербурге

Для того, чтобы удобнее было работать с объявлениями из Санкт-Петербурга, создадим соответствующую таблицу (`spb_data`) из основной.

```python
spb_data = good_data[good_data['locality_name'] == 'Санкт-Петербург']

# проверяем, все ли объявления из СПб вошли в новую таблицу
print(len(spb_data))
spb_data.info()
```

#### Расстояние до центра

Создадим новый столбец с расстоянием до центра в километрах (`km_to_center`) и построим график, отражающий зависимость средней цены объекта от расстояния.

```python
# сначала удалим из датасета строки с пропусками в столбце 'city_centers_nearest'
spb_data = spb_data.dropna(subset=['city_centers_nearest'])

spb_data['km_to_center'] = spb_data['city_centers_nearest'] / 1000

spb_data['km_to_center'] = spb_data['km_to_center'].astype(int)

# посмотрим, сработало ли преобразование
spb_data['km_to_center'].max()
```

Теперь построим график, который поможет определить, в каких пределах располагается престижный центр.

```python
(
    spb_data.pivot_table(index='km_to_center', values='last_price')
    .plot(grid=True, figsize=(10,5))
)
```

```python
spb_data.pivot_table(index='km_to_center', values='last_price', aggfunc=('mean', 'count'))
```

На графике обнаружился довольно чёткий момент резкого возрастания цены квартиры - он находится на 8 километре от центра. Таким образом, будем считать, что **престижный центр находится в радиусе 8 км**. 

Внимание привлёк странный пик на 27 км, однако стало ясно, что это выброс: в этом радиусе расположено всего два объекта. Видимо, и в радиусе 28 км также есть несколько дорогих объектов.
